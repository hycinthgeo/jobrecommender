# Section 4: Second inquiry to know more users' info & interest
* The inquiries include 6 aspects
<br> (a) If the user would like to match the "Minimum Qualifications" or "Preferred Qualifications" of the jobs
<br> (b) What's the user's highest degree?
<br> (c) What are the user's studied majors?
<br> (d) What are the users' interested experience/skills? 
<br> (e) What are the users' uninterested experience/skills? 
<br> (f) What is the similarity of the users' previous responsibilities, compared with this job?

* These 5 user's inputs will be used to further filter out unqualified jobs, and compute 3 scaled scores (0~1) to represent education, exp/skills, and responsibilities. 
<br> (a):         "min" -> match the minimum expectation from the employer; "preferred" -> match the ideal expectations from the employer
<br> (b):         will be used to filter out unqualified jobs
<br> (c):         will be used to compute major matching score. This system assumed a linear decreasing of employer's interest, when they are stating "BS/BA in major A, B, C, or D"
<br> (d) and (e): will be use to compute a skill matching score. 
<br> (f):         will be scaled to [0~1] to represent a similarity score for "Responsibilities"



```python
df, match_level, my_highest_edu, my_majors, my_skills, my_un_skills, my_responsibilities = rt.second_round_user_info_by_qa(df_min, df_pref)
```

```python
unlike_penalty = 1
df_w_scores = rt.get_featured_scores(df, match_level, my_highest_edu, my_majors, my_skills, my_un_skills, unlike_penalty, my_responsibilities)
df_w_scores[['Title','degree_matched','scaled_major_score','scaled_skill_score','scaled_resp_sim']].head()
```

# Section 5: Final ranking and result visualizaiton

```python
# compute final score and get the top 10
df_top10 = rt.compute_final_score_return_top10(df_w_scores, 1.0/3, 1.0/3, 1.0/3)
```

```python
#df_top10 = rt.compute_final_score_return_top10(df_w_scores, 1.0/3, 1.0/3, 1.0/3)
import matplotlib.pyplot as plt
import matplotlib.pylab as pl
import matplotlib.gridspec as gridspec
from mpl_toolkits.mplot3d import Axes3D

def plot_3D_results(xs, ys, zs, labels, titles, locs):
    fig = plt.figure(figsize = (18, 15))
    ax = fig.add_subplot(111, projection='3d')

    i = 0
    for j in range(len(xs)):
        for c, m in [('b', 'o')]:
            #print labels[i]
            s = 1.0/(labels[i] + 1) * 100
            ax.scatter(xs[j], ys[j], zs[j], c=c, marker=m, s= s)
            i += 1

    ax.set_xlabel('Scaled score 1: Resp. similarity', fontsize = 14)
    ax.set_ylabel('Scaled score 2: Skill match', fontsize = 14)
    ax.set_zlabel('Scaled score 3: Major match', fontsize = 14)

    ax.set_xlim([0, 1])
    ax.set_ylim([0, 1])
    ax.set_zlim([0, 1])
    
    #plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False
    #plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True

    #print plt.rcParams.keys()

    zdirs = [None for i in range(len(xs))]

    count = 0
    for zdir, x, y, z, label in zip(zdirs, xs, ys, zs, labels):
        label = '%3s' % (label)
        ax.text(x, y, z, label, zdir, fontsize = 12)
        count += 1
	plt.title('Top 10 Recommended Jobs', fontsize = 20, fontweight = 'bold', loc = 'left')
    

	legends = [None for i in range(len(titles))]
	for i in range(len(labels)):
		title = titles[i].split(",")
		legends[i] = "%3s - %s - %20s" %(str(labels[i]), title[0].strip(), locs[i])
	#print legends
    
    	#set legend positions
    box = ax.get_position()
    print [box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 0.9]
    ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 0.9])

    #ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), fancybox=True, shadow=True, ncol=1)
    
    plt.legend(legends)

    plt.show()
    fig.savefig(fig_dir_path + "Final_recommended_top10_jobs")


rtv.plot_3D_results(df_top10['scaled_resp_sim'], df_top10['scaled_skill_score'], df_top10['scaled_major_score'], df_top10.index, df_top10['Title'], df_top10['tag_loc'])  
```
